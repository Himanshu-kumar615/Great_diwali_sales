```python
# drop- delete unwanted columns, inplace-to save/ implement
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#Check for null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#Drop null values
df.dropna(inplace=True)
```

Exploratory Data Analysis

Gender

```python
df.columns
```

```python
A=sp.countplot(x='Gender',data=df)
```

```python
A = sp.countplot(x = 'Gender',data = df)

for bars in A.containers:
    A.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sp.barplot(x='Gender',y='Amount', data= sales_gen)
```

#**from the above graph we can see that most of the buyers are females and even the purchasing power of females are greater tha men.**

#***Age***

```python
df.columns
```

```python
sp.countplot(data=df, x='Age Group', hue='Gender')  
```

```python
A=sp.countplot(data=df, x='Age Group', hue='Gender')
for bars in A.containers:
    A.bar_label(bars)
```

```python
# Total Amount vs Age group
sales_age=df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
sp.barplot(data=sales_age, x='Age Group',y='Amount')
```

****From the above graph we see that most of the buyers are of age group between 26-35 yrs female****

```python
# Total number of orders from top 10 states

orders_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sp.set(rc={'figure.figsize':(15,5)})
sp.barplot(data=orders_state, x='State',y='Orders')
```

```python
# Total amount/sales from top 10 states

sales_state= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending= False).head(10)
plt.figure(figsize=(20,4))                                            # Fig size- (width,height)
sp.barplot(data=sales_state, x='State',y='Amount')
```

*****From above graphs we can see that most of the orders & total sales/amoungt are from Uttar Pradesh, Maharashtra, Karnataka respectively.*****

**Marital Status**

```python
plt.figure(figsize=(5,3)) 
A=sp.countplot(data=df, x='Marital_Status')
for bars in A.containers:
    A.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(5,4)) 
sp.barplot(data=sales_state, x='Marital_Status',y='Amount', hue='Gender')
```

**** From the above graphs we can see that most of the buyers are married(women) and they have high purschasing power.****

*****Occupation*****

```python
plt.figure(figsize=(20,7))
A=sp.countplot(data=df, x="Occupation")
for bars in A.containers:
    A.bar_label(bars)
```

*****from the above graph we can see that most of the buyers are working in IT,healthcare and Aviation sector*****

*****Product Category*****

```python
plt.figure(figsize=(22,5))
A=sp.countplot(data=df,x='Product_Category')
for bars in A.containers:
    A.bar_label(bars)
```

```python
sales_state= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(22,7))
sp.barplot(data=sales_state,x='Product_Category',y='Amount')
```

*****From above graph we can see taht most of the sold products are from food, Clothing and Electronic category*****

```python
sales=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
plt.figure(figsize=(15,5))
sp.barplot(data=sales,x='Product_ID',y='Orders')
```

```python
# Top 10 most sold products 
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

*****Conclusion*****

*****Married women age group 26-35 yrs from Up, Maharashtra & Karnataka working in IT, Healthcare & Aviation are more likely buy products from food, clothing & Electronic category*****
